```python
x = df.drop(columns=["Salary"], axis = 1)
y = df["Salary"]
```

```python
X_one_encoded = pd.get_dummies(
    x,
    columns=["Gender", "Education Level", "Job Title"],
    drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
numerical_columns = ["Age", "Years of Experience" , "Gender" , "Education Level", "Job Title"]
scaler = StandardScaler()
Xlable[numerical_columns] = scaler.fit_transform(Xlable[numerical_columns])


```

```python
Xlable[['Age' , 'Years of Experience' , 'Gender' , 'Education Level', 'Job Title']] = scaler.fit_transform(Xlable[['Age' , 'Years of Experience' , 'Gender' , 'Education Level', 'Job Title']])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_one_encoded, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train , y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
r2 = r2_score(y_test , y_predict)
r2
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
import numpy as np

# --- Adjusted R on whole dataset ---
y_pred_all = model.predict(X_one_encoded)
n, p = X_one_encoded.shape

r2_all = r2_score(y, y_pred_all)

n, p = x.shape
adjusted_r2_all = 1 - ((1 - r2_all) * (n - 1)) / (n - p - 1)

print("R (whole dataset):", r2_all)
print("Adjusted R (whole dataset):", adjusted_r2_all)

# --- Cross-validation R ---
r2_scores = cross_val_score(model, X_one_encoded, y, cv=5, scoring="r2")
print("Mean R from 5-fold CV:", np.mean(r2_scores))
print("All CV R scores:", r2_scores)
```
